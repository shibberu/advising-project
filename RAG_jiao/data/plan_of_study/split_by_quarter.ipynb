```python
with open("./data/plan_of_study.txt", 'r') as f, open("./data/plan_of_study_by_quarter.txt", 'w') as target:
    import re
    fall_regex = "Fall (.*)Winter"
    winter_regex = "Winter (.*)Spring" 
    spring_regex = "Spring (.*)"
    while True:
        l = f.readline()
        if l == None:
            break
        body = l.split("\"")[7] 
        source = l.split("\"")[3]
        print(source)
        fall_courses = re.search(fall_regex, body).group(1)
        winter_courses = re.search(winter_regex, body).group(1)
        spring_courses = re.search(spring_regex, body).group(1)
        injection = body.split("Fall")[0]
        
        segments_fall = injection.split(",")
        segments_fall[0] += " fall"
        segments_fall[1] += " in fall"
        new_injection_fall = ",".join(segments_fall)
    
        segments_winter = injection.split(",")
        segments_winter[0] += " winter"
        segments_winter[1] += " in winter"
        new_injection_winter = ",".join(segments_winter)
    
        segments_spring = injection.split(",")
        segments_spring[0] += " spring"
        segments_spring[1] += " in spring"
        new_injection_spring = ",".join(segments_spring)
    
        
        target.write(f'{{\"source\": \"{source}\", \"text\": \"{new_injection_fall}{fall_courses}\", \"lemma\": \"{new_injection_fall}{fall_courses}\"}}\n')
        target.write(f'{{\"source\": \"{source}\", \"text\": \"{new_injection_winter}{winter_courses}\", \"lemma\": \"{new_injection_winter}{winter_courses}\"}}\n')
        target.write(f'{{\"source\": \"{source}\", \"text\": \"{new_injection_spring}{spring_courses}\", \"lemma\": \"{new_injection_spring}{spring_courses}\"}}\n')
```
